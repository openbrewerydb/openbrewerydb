## Clean-up

```python
# Get rid of rows without `country`
df.dropna(subset=['country'], inplace=True)
```

```python
# Fill empty `created_at` and `updated_at` values with current-ish datetime
df.created_at.fillna('2020-10-23 00:00:00', inplace=True)
df.updated_at.fillna('2020-10-23 00:00:00', inplace=True)
```

```python
# Let's fix any rows Excel destroyed. We're going 
df.loc[df.created_at.str.match('1900'),'created_at'] = '2018-07-24 01:34:37'
```

```python
df.loc[df.updated_at.str.match('1900'),'updated_at'] = '2018-07-24 01:34:37'
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['updated_at'] = pd.to_datetime(df['updated_at'])
```

```python
df['city'] = df['city'].str.strip()
```
